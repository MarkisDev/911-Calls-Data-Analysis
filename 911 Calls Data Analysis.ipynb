# Data Analysis of 911 Calls

In this project, I have analyzed some 911 call data from [Kaggle](https://www.kaggle.com/mchirico/montcoalert). The data contains the following fields:

* lat : String variable, Latitude
* lng: String variable, Longitude
* desc: String variable, Description of the Emergency Call
* zip: String variable, Zipcode
* title: String variable, Title
* timeStamp: String variable, YYYY-MM-DD HH:MM:SS
* twp: String variable, Township
* addr: String variable, Address
* e: String variable, Dummy variable (always 1)

## Data and Setup

```python
#Read in the csv file as a dataframe called df
dataset = pd.read_csv('911.csv')
df=dataset.copy()
```

## Deatiled Analysis & Adding New Information

```python
#In the titles column there are "Reasons/Departments" specified before the title code. 
#These are EMS, Fire, and Traffic. 
#Creating a new column called "Reason" that contains this string value.

df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])
```

```python
#What is the most common Reason for a 911 call based off of this new column?
df['Reason'].value_counts()
```

```python
#Using seaborn to create a countplot of 911 calls by Reason.
sns.countplot(x=df['Reason'])
```

```python
#Based on time information, I have performed a couple of time related alaysis.
#What is the data type of the objects in the timeStamp column?
type(df['timeStamp'].iloc[0])
```

```python
#Converting timestamps in "String" to "TimeStamp" datatype i.e. a DataTime object.
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
#Now that the timestamp column are actually DateTime objects, we are creating 3 new columns called Hour, Month, and Day of Week. 
#You will create these columns based off of the timeStamp column

df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)
df['Month'] = df['timeStamp'].apply(lambda x: x.month)
df['Day of Week'] = df['timeStamp'].apply(lambda x: x.dayofweek)
```

```python
#Changing Numerical value Day of Week to Categorical values of Day names.

dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
#Using seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column.

sns.countplot(x="Day of Week", data=df,hue="Reason", palette="Greens")

#To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
#Using seaborn to create a countplot of the Month column with the hue based off of the Reason column.

sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
#On Analysis, we can see the above graph is missing some months! 9,10, and 11 are not there.
```

```python
#To see if we can maybe fill in this information,I'm plotting it in another way, 
#possibly a simple line plot that fills in the missing months.
```

```python
#Creating a gropuby object called byMonth,where you group the DataFrame by the month column.
#Using the count() method for aggregation.
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
#Create a simple plot off of the dataframe indicating the count of calls per month.
byMonth['twp'].plot()
```

```python
#Use seaborn's lmplot() to create a linear fit on the number of calls per month. 
#Need to reset the index to a column while plotting along the column used for groupBy.

sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
#Creating a new column called 'Date' that contains the date from the timeStamp column.
df['Date']=df['timeStamp'].apply(lambda t: t.date())
```

```python
#Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.**

df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

```python
#Recreating the same as 3 separate plots with each plot representing a Reason for the 911 call.
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
#Restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week.
#There are lots of ways to do this, but I have combined groupby with an unstack() method.
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
#Now create a HeatMap using this new DataFrame.

plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='coolwarm')
```

```python
#Creating a clustermap using this DataFrame.
sns.clustermap(dayHour,cmap='Blues')
```

```python
#Restructure the dataframe so that the columns become the Hours and the Index becomes the Month.
#There are lots of ways to do this, but I have combined groupby with an unstack() method.
```

```python
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
#Now create a HeatMap using this new DataFrame.

plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap="coolwarm")
```

```python
#Creating a clustermap using this DataFrame.
sns.clustermap(dayMonth,cmap='Blues')
```
